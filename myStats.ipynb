```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('trace_out.csv',header=None)
data = data.rename(columns={0: "pattern", 1: "frequency"})
data = data.sort_values('frequency', ascending=False)
```

```python
fig, ax = plt.subplots()

ax.bar(data['pattern'], data['frequency'], width=1)
plt.show()
```

```python
x = np.arange(0,data.shape[0],1)
y = np.array(data['frequency'])

plt.plot(x,y)
```
